# Random Number Generation
np.random includes some functions for random number generation. Below we try to generate a deterministic with added statistical noise from normal (Gaussian) distribution.

````python
import numpy as np
np.random?
````

````python
import numpy as np

n = 200

# linespace produces equi-distant x values for, is useful for plotting curves
x = np.linspace(0.0, 2.0, n)

# initiate parameters as a habit
mu, sigma = 0, 50

#set seeed to regenerate the same random numbers, 
#useful for reproducible results.
np.random.seed(1)


e = np.random.normal(loc=mu, scale= sigma, size=len(x))

y = np.exp(3*x)+e 




````

````python
import matplotlib.pyplot as plt

#this command is required for jupyter notebook, 
#otherwise everytime you must use plt.show()

%matplotlib inline


# scatter plot of data
plt.plot(x, y,".",mew=3);

# add a red curve
plt.plot(x, np.exp(3*x),"--r",lw=3);

# set axis labels
plt.xlabel('Explanatory variable (x)');
plt.ylabel('Dependent variable (y)');
````

The red curve is defined by the function 

\begin{equation}
 \\  f(x_i) = e^{\ 3 x_i}, i=1,\ldots, n \ \ ,
\end{equation}

and the blue dots are actually the dependent variable defined by 

\begin{equation}
    y_i = e^{\ 3 x_i} + \varepsilon_i \ \ \text{where}  \ \  \varepsilon_i \sim \mathcal{N}(0,50^2) \ \ .
\end{equation}
